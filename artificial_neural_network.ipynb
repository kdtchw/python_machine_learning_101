```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_csv("./data/neural_network/Churn_Modelling.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,2] = le.fit_transform(X[:,2])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [1])], remainder="passthrough")
X = np.array(ct.fit_transform(X))
```

```python
X[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# initialize the ann
ann = tf.keras.models.Sequential()
```

```python
# add an input layer and 1st hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
```

```python
# add a 2nd hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation="relu"))
```

```python
# add an output layer
ann.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# compile the ann
ann.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# train the ann on the training dataset
ann.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
# predict whether customer will leave if geography=France, credit_score=600, gender=male, age=40, tenure=3,
# balance=60000, number_of_products=2, has_credit_card=yes, is_active_member=yes, estimated_salary=50000
ann.predict(sc.transform([[1, 0, 0, 600, 1, 40, 3, 60000, 2, 1, 1, 50000]]))
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
np.set_printoptions(threshold=len(X_train))
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
accuracy_score(y_test, y_pred)
```
